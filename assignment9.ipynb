Matplotlib is a Python library used for creating high-quality static plots, interactive plots, and animations. It is built on top of NumPy arrays and designed to work with the broader SciPy stack.The Pyplot module of Matplotlib is used to create a variety of plots. Here are five plots that can be plotted using the Pyplot module of Matplotlib:
1.Line Plot: A line plot is a type of plot that displays information as a series of data points connected by straight line segments. It is useful for visualizing trends in data over time 
2.Histogram: A histogram is a type of plot that displays the distribution of a dataset. It is useful for visualizing the shape of the data and identifying outliers 
3.Scatter: A scatter plot is a type of plot that displays the relationship between two variables. It is useful for identifying patterns and trends in data
4.3D Plot: A 3D plot is a type of plot that displays information in three dimensions. It is useful for visualizing complex data and identifying patterns that may not be visible in two dimensions 
5.Image: An image plot is a type of plot that displays an image. It is useful for visualizing images and identifying patterns in them

Q2

A scatter plot is a type of plot that displays the relationship between two variables. It is useful for identifying patterns and trends in data 

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot of X and Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Q3

The subplot() function in Matplotlib is used to create multiple plots within a single figure. It is useful when we want to display multiple plots side-by-side or stacked on top of each other

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2, 2)
fig.suptitle('Four Line Plots')

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

plt.show()
```

A bar plot is a type of plot that displays the distribution of a dataset. It is useful for visualizing the shape of the data and identifying outliers. A bar plot can be plotted horizontally or vertically

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title('Bar Plot of Company vs Profit')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title('Horizontal Bar Plot of Company vs Profit')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

A box plot is a type of plot that displays the distribution of a dataset. It is useful for visualizing the shape of the data and identifying outliers. A box plot can be used to display the median, quartiles, and range of a dataset

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title('Box Plot of Box1 and Box2')
plt.xlabel('Box')
plt.ylabel('Value')
plt.show()
```
